```python
import os
import random
import shutil

# ścieżki
base_dir = "/Users/sebastianstarzec/Downloads/archive/"
images_dir = os.path.join(base_dir, "images")
ann_dir = os.path.join(base_dir, "annotations")

# nowe foldery
output_dir = "dataset"
train_img = os.path.join(output_dir, "images/train")
val_img = os.path.join(output_dir, "images/val")
train_ann = os.path.join(output_dir, "annotations/train")
val_ann = os.path.join(output_dir, "annotations/val")

# tworzenie folderów
for path in [train_img, val_img, train_ann, val_ann]:
    os.makedirs(path, exist_ok=True)

# lista obrazów
images = [f for f in os.listdir(images_dir) if f.endswith((".jpg", ".png"))]

# mieszanie
random.shuffle(images)

# podział
split_idx = int(0.8 * len(images))
train_files = images[:split_idx]
val_files = images[split_idx:]

def move_files(file_list, img_dest, ann_dest):
    for file in file_list:
        name = os.path.splitext(file)[0]
        
        img_src = os.path.join(images_dir, file)
        ann_src = os.path.join(ann_dir, name + ".xml")
        
        shutil.copy(img_src, os.path.join(img_dest, file))
        
        if os.path.exists(ann_src):
            shutil.copy(ann_src, os.path.join(ann_dest, name + ".xml"))

# kopiowanie
move_files(train_files, train_img, train_ann)
move_files(val_files, val_img, val_ann)

print("✅ Podział zakończony!")
```
